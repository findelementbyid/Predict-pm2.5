```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('./新竹_2019.csv', encoding = 'big5')
data = data[4825:].reset_index(drop = True)
data.columns = data.columns.str.strip()
data.drop(['日期', '測站', '測項'], axis=1, inplace = True)
data
```

```python
for index, row in data.iterrows():
    for i in range(len(row)):
        if '#' in row[i] or '*' in row[i] or 'A' in row[i] or 'x' in row[i]:
            row[i] = np.nan
```

```python
new_data = pd.DataFrame(data, dtype = float)
new_data = new_data.interpolate(method='linear', axis = 1)
new_data = new_data.interpolate(method='linear', axis = 0)
```

```python
train = new_data[:1098]
test = new_data[1098:].reset_index(drop = True)
```

```python
train
```

```python
test
```

```python
new_train = train.iloc[0:18]

for i in range(60):
    d = train.iloc[0 : 18]
    new_train = pd.concat([new_train, d.set_index(new_train.index)], axis = 1)
    train = train.drop(d.index, axis = 0)

new_train
```

```python
new_test = test.iloc[0:18]

for i in range(30):
    d = test.iloc[0 : 18]
    new_test = pd.concat([new_test, d.set_index(new_test.index)], axis = 1)
    test = test.drop(d.index, axis = 0)

new_test
```

```python
# a-1.未來第一個小時 in training data set

first_train_x = []
first_train_y = []

for i in range(1458):
    first_train_y.append(float(new_train.iloc[9, i + 6]))
    tempArr = []
    for j in range(6):
        temp = (float(new_train.iloc[9, i + j]))
        tempArr.append(temp)
        #x[i].append(temp)
    first_train_x.append(tempArr)
```

```python
# Traing linear regression model

first_lireg = LinearRegression().fit(first_train_x, first_train_y)
first_lireg.score(first_train_x, first_train_y)
```

```python
# a-1.未來第一個小時 in testing data set

first_test_x = []
first_test_y = []

for i in range(738):
    first_test_y.append(float(new_train.iloc[9, i + 6]))
    tempArr = []
    for j in range(6):
        temp = (float(new_train.iloc[9, i + j]))
        tempArr.append(temp)
        #x[i].append(temp)
    first_test_x.append(tempArr)
```

```python
# Linear Regression MAE for 未來第一小時

first_lireg_y_pred = first_lireg.predict(first_test_x)
mean_absolute_error(first_test_y, first_lireg_y_pred)
```

```python
# Random Forest Regression MAE for 未來第一小時

first_rfger = RandomForestRegressor(max_depth = 2, random_state = 0)
first_rfger.fit(first_train_x, first_train_y)
first_rfger_y_pred = first_rfger.predict(first_test_x)
mean_absolute_error(first_test_y, first_rfger_y_pred)
```
